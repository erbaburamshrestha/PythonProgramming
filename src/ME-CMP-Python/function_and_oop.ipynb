```python
# whether x is even or odd
def evenOdd(x):
    if (x % 2 == 0):
        print("even")
    else:
        print("odd")
# Driver code to call the function
evenOdd(2)
evenOdd(3)
```

```python
def is_prime(n):
    if n in [2, 3]:
        return True
    if (n == 1) or (n % 2 == 0):
        return False
    r = 3
    while r * r <= n:
        if n % r == 0:
            return False
        r += 2
    return True
print(is_prime(78), is_prime(79))
```

```python
# *args for variable number of arguments
def myFun(*argv):
    for arg in argv:
        print(arg)
# Driver code
myFun('Hello', 'I', 'am', 'Baburam', 'Shrestha') 
```

```python
# *kwargs for variable number of keyword arguments
def myFun(**kwargs):
    for key, value in kwargs.items():
        print("%s == %s" % (key, value)) 
# Driver code
myFun(first='Babu', mid='Ram', last='Shrestha')  
```

# Recursive Functions in Python

```python
def factorial(n):
    if n == 0:  
        return 1
    else:
        return n * factorial(n - 1) 
print(factorial(4))
```

# Anonymous/Lambda Functions in Python


```python
# Example 1:
def cube(x): 
    return x*x*x
print(cube(7))
cube_v2 = lambda x : x*x*x
print(cube_v2(7))

```

```python
# Example 2:
add = lambda a, b: a + b
print(add(3, 7))
```

```python
# Example 3:
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

# List/Tuple/Dictionary in Functions

```python
def sum_elements(input_list):
    sum = 0
    for x in input_list:
        sum += x 
        return sum

# Example usage:
original_list = [1, 2, 3, 4, 5]
result = sum_elements(original_list)
print("Original List:", original_list)
print("Sum of Elements in List:", result)
```

```python
def find_multiply(input_tuple):
    mul = 1
    for elem in input_tuple:
        mul *= elem
        return mul
# Example usage
data = (1, 2, 3, 4, 5)
result = find_multiply(data)
print("Input Tuple:", data)
print("Multiply of elements in Tuple:", result)
```

```python
def filter_passed_students(grades_dict):
    passed_students = {} 
    for student, grade in grades_dict.items():
        if grade >= 50: 
            passed_students[student] = grade 
    return passed_students
student_grades = {
    'Alice': 85,
    'Bob': 45,
    'Charlie': 70,
    'David': 30,
    'Eva': 90
}

passed_students= filter_passed_students(student_grades)
print(passed_students)
```

# Object Oriented Programming (OOP)

```python
class Dog:
       attr1 = "mammal" # Class attribute

# Driver code
rodger = Dog() # Object instantiation
rodger.name = "Rodger" 

tommy = Dog()
tommy.name = "Tommy"

# Accessing class attributes
print("Rodger is a {}".format(rodger.__class__.attr1))
print("Tommy is also a {}".format(tommy.__class__.attr1))

# Accessing instance attributes
print("My name is {}".format(rodger.name))
print("My name is {}".format(tommy.name))
```

```python
class Dog:
    attr1 = "mammal" # Class attribute
    def __init__(self, name):
        self.name=name # class instance attributes
    def speak(self):
        print('f Dog name: {self.name}')
# Driver code
rodger = Dog("Rodger" ) # Object instantiation
tommy = Dog("Tommy")
# Accessing class methods
rodger.speak()
tommy.speak()
# Accessing instance attributes
print("My name is {}".format(rodger.name))
print("My name is {}".format(tommy.name))
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
emp1 = Employee("John Doe", 5000)
emp2 = Employee("Jane Smith", 6000)
print(emp1.name, emp1.salary)
```
