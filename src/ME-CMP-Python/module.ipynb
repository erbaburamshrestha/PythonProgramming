```python
# importing sqrt() and factorial from the module math
from math import sqrt, factorial
# if we simply do "import math", then
# math.sqrt(16) and math.factorial() # are required.
print(sqrt(16))
print(factorial(6))
```

```python
# importing sqrt() and factorial from the module math
import math as mt
# if we simply do "import math", then
# math.sqrt(16) and math.factorial() are required.
print(mt.sqrt(16))
print(mt.factorial(6))
```
