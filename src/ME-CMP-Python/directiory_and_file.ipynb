# Directory

```python
import os
```

```python
# Create a single directory
os.mkdir('my_directory')
# Create nested directories
os.makedirs('parent_directory/child_directory')
```

```python
# Listing Files in a Directory
files = os.listdir('my_directory')
print(files)
```

```python
# Changing the Current Working Directory
os.chdir('parent_directory')
print("Current Working Directory:", os.getcwd())
```

```python
os.chdir('../my_directory')
print("Current Working Directory:", os.getcwd())
```

```python
os.chdir('../')
```

```python
# Deleting a Directory
os.rmdir('my_directory') # Remove an empty directory
os.removedirs('parent_directory/child_directory') # Remove nested directories
```

```python
# Importing the pathlib Module
from pathlib import Path
```

```python
# Creating a Directory
# Create a single directory
Path('my_directory').mkdir(exist_ok=True)
# Create nested directories
Path('parent_directory/child_directory').mkdir(parents=True, exist_ok=True)
```

```python
# Listing Files in a Directory
for file in Path('.').iterdir():
    print(file)
```

```python
# Changing the Current Working Directory
current_path = Path.cwd()
print("Current Working Directory:", current_path)
```

```python
# Deleting a Directory
import shutil
# Remove an empty directory
Path('my_directory').rmdir()
# Remove a non-empty directory
shutil.rmtree('parent_directory')
```

# File

```python
# Importing the Required Modules:Before you can perform file operations, you need to import the necessary modules:
import os
import shutil
```

```python
# Create and write to a file
with open('old_filename.txt', 'w') as f:
    f.write('This is an example file\n')
    f.write('Created using Python file operations')

# Verify the file was created
print("File created:", Path('old_filename.txt').exists())
```

```python
# Renaming a File: To rename a file, you can use the os.rename() function.
# Renaming a file
os.rename('old_filename.txt', 'new_filename.txt')
```

```python
# Moving a file
shutil.move('new_filename.txt', '../filename.txt')
```

```python
# Copying a File
# You can copy a file using the shutil.copy() or shutil.copy2() functions. 
# The shutil.copy2() also copies the fileâ€™s metadata.
shutil.copy('example.txt', './filename_copy.txt')
# Using copy2 for metadata
shutil.copy2('example.txt', './filename_copy2.txt')
```

```python
# Removing a file
os.remove('./filename_copy2.txt')
```

```python
# Removing an empty directory
os.rmdir('file')
# Removing a directory and its contents
shutil.rmtree('files')
```

```python
with open('filename.txt', 'w') as f:
    f.write('This is an example file\n')
    f.write('Created using Python file operations')
    f.write('This is an example file\n')
    f.write('Created using Python file operations')
    f.close()
```
