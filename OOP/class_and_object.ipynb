OOP in Python objects and class
4 Pillers of OOP
    - Data Abstraction
    - Encapsulation
        bundling of attributes and methods inside a single class
        achieve data hiding
        private attributes using underscore ( _ or __ )
    - Inheritance
        way of creating a new class for using details of an existing class without modifying it
        newly formed class is a derived class (or child class).
        the existing class is a base class(or parent class)
    - Polymorphism
        more than one form
        same function in different class
        
An object is any entity that has 
    - attributes
    - behaviors

A parrot is an object. It has
    attributes - name, age, color, etc.
    behavior - dancing, singing, etc.

```python
class Animal:
    #list of attributes
    name = "baburam"
    age = 45
    address = "Kathmandu"
    color = "black"
    weight = 70
    height = 5.7
    #list of methods
    def eat(self):
        print(f"{self.name} is eating.")
    def sleep(self):
        print(f"{self.name} is sleeping.")
    def details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Address: {self.address}")
        print(f"Color: {self.color}")
        print(f"Weight: {self.weight} kg")
        print(f"Height: {self.height} ft")
#creating object of the class
anim = Animal()
#accessing attributes
print(anim.name)
print(anim.age)
#accessing methods
anim.eat()
anim.sleep()
anim.details()
```

```python
class Animal:
    name = "Rocky"  #public attributes
    _weight = 50    #protected attribute
    __height = 5.5  #private attributes
    def get_name_weight_height(self):
        print(f"Name: {self.name}")
        print(f"Weight: {self._weight} kg")
        print(f"Height: {self.__height} ft") 

#creating object of the class
anim = Animal()
#accessing methods
anim.get_name_weight_height() # from function within the class can access all types of attributes
print(anim.name)              #public attribute can be accessed outside the class
print(anim._weight)           #protected attribute can be accessed outside the class but not recommended
# print(anim.__height)          #private attribute cannot be accessed outside the  defined class
```

```python
class Animal:
    name = "Rocky"  #public attributes
    _weight = 50    #protected attribute
    __height = 5.5  #private attributes
    def get_name_weight_height(self):
        print(f"Name: {self.name}")
        print(f"Weight: {self._weight} kg")
        print(f"Height: {self.__height} ft") 

class Dog(Animal): # Dog is child(subclass) and Animal is parent(superclass), Dog inherits from Animal
    def bark(self, name):
        print(f"{name} Woof! Woof!")
#creating object of the class
dg = Dog()
print(dg.name) # accessing public attribute from parent class
print(dg._weight) # accessing protected attribute from parent class
dg.get_name_weight_height()
```

```python
class Animal:
    name = "Rocky"  #public attributes
    _weight = 50    #protected attribute
    __height = 5.5  #private attributes
    def get_name_weight_height(self):
        print(f"Name: {self.name}")
        print(f"Weight: {self._weight} kg")
        print(f"Height: {self.__height} ft") 

class Dog(Animal): # Dog is child(subclass) and Animal is parent(superclass), Dog inherits from Animal
    def bark(self, name):
        print(f"{name} Woof! Woof!")
    def get_weight(self, weight):
        print(f"Weight: {weight} kg")
    def get_weights(self):
        print(f"Weight: {self._weight} kg")
    def get_name(self):
        self.get_name_weight_height()
#creating object of the class
dg = Dog()
dg.bark("Tom")
dg.bark(dg.name) # accessing public attribute from parent class
dg.get_weight(dg._weight) # accessing protected attribute from parent class
dg.get_weights() # accessing protected attribute from parent class via method
dg.get_name() # accessing all attributes from parent class via method
```

```python
class Animal:
    name = "Rocky"
    def get_name(self):
        print(f"Name: {self.name}...........")

class Dog(Animal):
    def get_name(self, age):
        print(f"Name: {self.name} and Age: {age}")
        print(f"Name: {self.name}")
        print(f"Name: {self.name}")
ani = Animal()
ani.get_name()
dg = Dog()
dg.get_name(5) # function overriding
```

```python
class Mathematics:
    def add(self, a, b):
        return a + b

math = Mathematics()
print(math.add(10, 5)) # normal addition
print(math.add("10", "5")) # it does concatenation
# function overloading
```
